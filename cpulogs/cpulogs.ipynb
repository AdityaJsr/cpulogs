```
paths = ["/FileStore/tables/CpuLogData2019_09_16.csv", "/FileStore/tables/CpuLogData2019_09_17.csv", "/FileStore/tables/CpuLogData2019_09_18.csv","/FileStore/tables/CpuLogData2019_09_19.csv","/FileStore/tables/CpuLogData2019_09_20.csv","/FileStore/tables/CpuLogData2019_09_21.csv"]

df = spark.read.csv(paths, header="true", inferSchema="true")
```

```
df = df.select("DateTime","keyboard","mouse","user_name")
display(df)
```

```
from pyspark.sql import functions as F
from pyspark.sql import *

split_col = F.split(df['DateTime'], ' ')
df = df.withColumn('Date', split_col.getItem(0))
df = df.withColumn('Time', split_col.getItem(1))
display(df)
```

```
# Creating table
new_temp_table_name = "CpuLogData"
df.createOrReplaceTempView(new_temp_table_name)
```

```
# Finding users with highest number of average hours 
highest_working_hours = sqlContext.sql("SELECT user_name,from_unixtime(round((((count(user_name)*5)*60)/6),2),'HH:mm') active_hours FROM CpuLogData WHERE keyboard !=0 AND mouse !=0 GROUP BY user_name ORDER BY active_hours desc")
display(highest_working_hours)
```

```
# Finding users with lowest number of average hours 
lowest_working_hours = sqlContext.sql("SELECT user_name,from_unixtime(round((((count(user_name)*5)*60)/6),2),'HH:mm') active_hours FROM CpuLogData WHERE keyboard !=0 AND mouse !=0 GROUP BY user_name ORDER BY active_hours asc")
display(lowest_working_hours)
```

```
# Finding users with highest numbers of idle hours
highest_idle_hours = sqlContext.sql("SELECT user_name,from_unixtime(round((((count(user_name)*5)*60)/6),2),'HH:mm') idle_hour FROM CpuLogData WHERE keyboard == 0 AND mouse ==0 GROUP BY user_name ORDER BY idle_hour desc")
display(highest_idle_hours)
```

```
all_query = sqlContext.sql("select * from CpuLogData")
display(all_query)
```

```
#User active time 
user_login_time = sqlContext.sql("SELECT user_name,Date,to_timestamp(Min(DateTime)) login_time From CpuLogData WHERE keyboard !=0 OR mouse !=0 GROUP BY user_name,Date ORDER BY Date")
temp_table_name = "user_login_time"
user_login_time.createOrReplaceTempView(temp_table_name)
display(user_login_time)
```

```
split_col = F.split(user_login_time['login_time'], ' ')
user_login_time = user_login_time.withColumn('Time', split_col.getItem(1))
display(user_login_time)
```
